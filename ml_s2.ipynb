```
arr1=np.array([[1,2],[3,4]])
arr2=np.array([[4,5],[6,7]])
arr=np.concatenate((arr1,arr2))
print(arr)
```

```
#stacking along rows
#vstack()
#hstack()
arr1=np.array([1,2,3])
arr2=np.array([4,5,6])
arr=np.hstack((arr1,arr2))
print(arr)
```

```
arr1=np.array([[1,2],[3,4]])
arr2=np.array([[4,5],[6,7]])
arr=np.vstack((arr1,arr2))
print(arr)
```
