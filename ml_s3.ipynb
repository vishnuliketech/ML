```
'''
Intro to panda?(pan<el> da<ta>)
Data Frame(2D) and series(1D)
1. data processing and data analysis 
2. manage large dataset
3. pandas is fast than numpy

Diff numpy and pandas
1.numpy array matrix       1.pandas Series and DataFrame
2.Less memory              2.More memory
3.numerical data           3.Tabular data
4.multi-dimensional        4.2D table object (DATA FRAME)
'''

import pandas as pd
import numpy as np
```

```
s = pd.Series()
print(s)
```

```
data = np.array(['a','b','c'])
s = pd.Series(data)
print(s)
```

```
order = [1,2,3]
s1 = pd.Series(data,index=order)
print(s1)
```

```
#create a series from dictionary
data = {'a':0,
        'b':1,
        'c':2,}
s2 = pd.Series(data)
print(s2)
```

```
#create a series from dictionary
data = {'a':0,
        'b':1,
        'c':2,}
s3 = pd.Series(data,index=['b','c','d'])
print(s3)
```

```
#How to access data from Series with position
s = pd.Series([1,2,3,4,5],index=['a','b','c','d','e'])
print(s)
```

```
df = pd.DataFrame()
print(df)
```

```
data1 = [1,2,3,4,5]
df = pd.DataFrame(data1)
print(df)
```

```
data1 = [['alex',10],['raj',20],['shrin',30]]
df = pd.DataFrame(data1,columns = ['NAME','AGE'])
print(df)
```

```
data2 = {'NAME':['vijay','vishnu','vishal'],'AGE':[23,67,90]}
df = pd.DataFrame(data2,index=[1,2,3])
print(df)
```

```
#createing dataframe from list of dictionary
data4 = [{'a':1,'b':2,'c':3},{'a':56,'b':56,}]
df = pd.DataFrame(data4)
print(df)
```
