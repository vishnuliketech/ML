```
#head and tail
#head()
#tail()
print("The original series is    : ",s)
print("The actual data series is : ",s.values)
print("The first two rows        : ",s.head(4))
print("The last  two rows        : ",s.tail(2))
```

```
#Create a dictionary of series 
d = {'name':pd.Series(['tom','james','ricky','vivin','steve','smith','jack']),
     'Age':pd.Series([25,56,34,56,78,23,50]),
     'rating':pd.Series([3,4,5,2,6,77,8])}
df=pd.DataFrame(d)
print(df)
```

```
# Transpose will interchange rows and columns
print("The transpose of the data series is :")
print(df.T)
```

```
print("Row axis label")
print(df.axes)
```

```
print("Row axis labels and column axis labels: ")
print(df.dtypes)
```

```
#axis=1
print(df.sum(1))
```

```
print(df.mean())
```

```
# ******very******* The describe() function computes a summary of statics pretaining to the data 
print(df.describe())
```

```
#applying aggregation on dataframe
df=pd.DataFrame(np.random.randn(10,4),index=pd.date_range('1/1/2000',periods=10),columns=['A','B','C','D'])
print(df)
```

```
df=pd.DataFrame(np.random.randn(10,4),index=pd.date_range('1/1/2000',periods=10),columns=['A','B','C','D'])
print(df)
#rollint()
r=df.rolling(window=3,min_periods=1)
print(r)
```
